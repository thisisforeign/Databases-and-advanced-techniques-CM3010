```python
import csv

undernourish_path = './midterm/dataset/prevalence-of-undernourishment.csv'
gdp_path = './midterm/dataset/South_Asian_dataset.csv'

unique_entity = set()
unique_country = set()
unique_year_undernourish = set()
unique_year_gdp = set()
with open(undernourish_path, mode='r') as f1:
    csvFile = csv.reader(f1)
    header = next(csvFile)
    for lines in csvFile:
        unique_entity.add(lines[0])
        unique_year_undernourish.add(lines[1])
        
    with open(gdp_path, mode='r') as f2:
        csvFile2 = csv.reader(f2)
        for lines in csvFile2:
            unique_country.add(lines[0])
            unique_year_gdp.add(lines[1])
        
        same_countries = unique_entity.intersection(unique_country)
        same_year = unique_year_undernourish.intersection(unique_year_gdp)
```

```python
with open(undernourish_path, mode='r') as f1:
    csvFile = csv.reader(f1)
    
    with open('undernourishment-updated.csv', mode='w', newline='') as output:
        csv_write = csv.writer(output)

        header = next(csvFile)
        csv_write.writerow(header)
        
        for lines in csvFile:
            if lines[0] in same_countries:
                if lines[1] in same_year:
                    csv_write.writerow(lines)
```

```python
with open(gdp_path, mode='r') as f2:
    csvFile2 = csv.reader(f2)

    with open('gdp-updated.csv', mode='w', newline='') as output:
        csv_write = csv.writer(output)

        header = next(csvFile2)
        csv_write.writerow(header)
        
        for lines in csvFile2:
            if lines[0] in same_countries:
                if lines[1] in same_year:
                    csv_write.writerow(lines[:3])
```

START OF CLEAN

```python
to_clean_gdp = []
specific_columns = [0,1,2,5,6,8,9]
with open(gdp_path, mode='r') as f1:
    csvFile = csv.reader(f1)
    
    for lines in csvFile:
        if lines[0] in same_countries:
            if lines[1] in same_year:
                add_To = [lines[i] for i in specific_columns]
                to_clean_gdp.append(add_To)
```

```python
sorted_by_country = sorted(to_clean_gdp, key=lambda x: x[0])
gdp_growth = []

for i in range(0, len(sorted_by_country), 23):
    one_country = sorted_by_country[i:i+23]
    gdp_growth.append(0)
    for j in range(22):
        if ',' in one_country[j+1][2]:
            final = float(one_country[j+1][2].replace(',','.'))
        else:
            final = int(one_country[j+1][2])
        if ',' in one_country[j][2]:
            initial = float(one_country[j][2].replace(',','.'))
        else:
            initial = int(one_country[j][2])
        gdp_growth.append((final - initial) / initial * 100)
```

```python
gdp_per_capita = []

for i in range(len(sorted_by_country)):
    if ',' in sorted_by_country[i][2]:
        gdp = float(sorted_by_country[i][2].replace(',','.'))
    else:
        gdp = int(sorted_by_country[i][2])
    population = int(sorted_by_country[i][4])
    gdp_per_capita.append(gdp / population)
```

```python
pop_growth = []

for i in range(0, len(sorted_by_country), 23):
    one_country = sorted_by_country[i:i+23]
    pop_growth.append(0)
    for j in range(22):
        final = int(one_country[j+1][4])
        initial = int(one_country[j][4])
        pop_growth.append((final - initial) / initial * 100)
```

```python
sorted_by_country
```

```python
for row in sorted_by_country:
    if ',' in row[2]:
        row[2] = float(row[2].replace(',', '.'))
```

```python
for i in range(len(gdp_growth)):
    sorted_by_country[i].insert(3, gdp_growth[i])
    sorted_by_country[i].insert(4, gdp_per_capita[i])
    sorted_by_country[i].insert(7, pop_growth[i])
sorted_by_country
```

```python
sorted_by_year = sorted(sorted_by_country, key=lambda x: int(x[1]))
sorted_by_year
```

```python
with open(gdp_path, mode='r') as f2:
    csvFile2 = csv.reader(f2)

    with open('gdp-updated.csv', mode='w', newline='') as output:
        csv_write = csv.writer(output)

        header = next(csvFile2)
        csv_write.writerow(header)
        
        for lines in sorted_by_year:
            if lines[0] in same_countries:
                if lines[1] in same_year:
                    csv_write.writerow(lines)
```
